In this analysis we are analyzing eBay Kleinanzeigan data from Germany. The aim is to clean and learn more about the dataset.

```python
import numpy as np
import pandas as pd
from pandas import *
```

```python
autos = pd.read_csv(r'C:\Users\bbeckenb\OneDrive\Documents\Local Datasets\autos.csv', encoding="Windows-1252")
```

```python
autos.rename(columns={"yearOfRegistration":"registration_year", "lastSeen":"last_seen","dateCrawled":"date_crawled", "monthOfRegistration":"registration_month", "notRepairedDamage":"unrepaired_damage","dateCreated":"ad_created", "kilometer": "odometer_km"}, inplace=True)
autos.columns
autos.info()
autos.head()
    
```

```python
autos.describe(include='all')
```

```python
autos.drop(columns=['seller', 'offerType'])
```

```python
print(autos.loc[:, 'odometer_km'].min(), autos.loc[:, 'odometer_km'].max())
autos.loc[:, 'odometer_km'].value_counts()
```

```python
price = autos.loc[:, 'price']
print(price.min(), price.max())
#price.value_counts()
price.sort_index(ascending=False)
price.describe()
```

```python
autos.loc[price > 100000, 'price'] = 0
autos.drop(autos[autos['price'] < 100].index, inplace=True)
price = autos.loc[:, 'price']
price.describe()

odometer_km = autos.loc[:, 'odometer_km']
odometer_km.value_counts()
```

```python
autos[['date_crawled','ad_created','last_seen']][0:5]
autos['date_crawled'] = autos['date_crawled'].str[:10]
autos.loc[:, 'date_crawled'].value_counts(normalize=True, dropna=False).sort_index()
```

```python
autos['ad_created'] = autos['ad_created'].str[:10]
autos.loc[:, 'ad_created'].value_counts(normalize=True, dropna=False).sort_index()
```

```python
autos['last_seen'] = autos['last_seen'].str[:10]
autos.loc[:, 'last_seen'].value_counts(normalize=True, dropna=False).sort_index()
```

```python
autos['registration_year'].describe()
```

```python
autos.loc[(autos.loc[:, 'registration_year'] < 1900) | (autos.loc[:, 'registration_year'] > 2016), 'registration_year'] = np.nan
autos['registration_year'].describe()
```

```python
brand = autos.loc[:, 'brand']
brand.value_counts()
```

```python
brand_unique = brand.unique()
#print(brand_unique)
avg_price_by_brand = {}
for car_type in brand_unique:
    avg_price_by_brand[car_type] = autos.loc[autos.loc[:, 'brand'] == car_type, 'price'].mean()
    
for row in avg_price_by_brand:
    print(row, ':', round(avg_price_by_brand[row], 2))
```

```python
brand_unique = brand.unique()
#print(brand_unique)
avg_km_by_brand = {}
for car_type in brand_unique:
    avg_km_by_brand[car_type] = autos.loc[autos.loc[:, 'brand'] == car_type, 'odometer_km'].mean()
    
for row in avg_km_by_brand:
    print(row, ':', round(avg_km_by_brand[row], 2))
```

```python
brand_avg_price_series = pd.Series(avg_price_by_brand)
brand_avg_km_series = pd.Series(avg_km_by_brand)
df = pd.DataFrame(brand_avg_price_series, columns=['mean_price'])
df['mean_km'] = brand_avg_km_series
df.sort_values('mean_price', ascending=False)
```
